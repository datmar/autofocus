```python
files = np.asarray(glob.glob('../../../data/lpz_2012-2014/lpz_2012-2014/raw/**/*.JPG', recursive=True))
```

```python
file_subset = np.random.choice(files, size=16)
processed = np.asarray(glob.glob('../../../data/lpz_2012-2014/lpz_2012-2014/processed/images/*.JPG', recursive=True))
```

```python
plt.figure(figsize=(12,12))
nrow = 4
ncol = 4
fig, axs = plt.subplots(nrows=nrow, ncols=ncol, figsize=(12,12))
for i, ax in enumerate(axs.reshape(-1)):
    ax.set_xlabel(str(i+1))
    ax.imshow(cv.imread(file_subset[i])[:,:,::-1])
```
